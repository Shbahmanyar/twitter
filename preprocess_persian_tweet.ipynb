```python
import pandas as pd
import re
from hazm import *

# Mount Google Drive
#from google.colab import drive
#drive.mount('/content/drive')

# Load Persian stop words
stop_words = []
with open('/content/drive/MyDrive/persian_stop_words.txt', 'r') as file:
    stop_words = file.read().split('\n')

```

```python
df = pd.read_csv('/content/drive/MyDrive/mn/ok/final_4407.csv',encoding="utf-8")
```

```python
import regex as re

# Remove URLs
df['preprocess'] = df['Text'].apply(lambda text: re.sub(r'http\S+|www\S+', '', text))

# Remove punctuation marks and English numbers
df['preprocess'] = df['preprocess'].apply(lambda text: re.sub(r'[a-zA-Z0-9]+', '', text))
df['preprocess'] = df['preprocess'].apply(lambda text: re.sub(r'[^\w\s0-9a-zA-Z]+', '', text))

# Remove mentions of user screen names
df['preprocess'] = df['preprocess'].apply(lambda text: re.sub(r'@\w+', '', text))

# Remove underscores
df['preprocess'] = df['preprocess'].apply(lambda text: text.replace('_', ''))

# Normalize the text using Hazm library
normalizer = Normalizer()
df['preprocess'] = df['preprocess'].apply(lambda text: normalizer.normalize(text))


# Remove stop words
normalizer = Normalizer()
tokenizer = WordTokenizer()
df['preprocess'] = df['preprocess'].apply(lambda text: ' '.join(word for word in tokenizer.tokenize(text) if word not in stop_words))

# Save the preprocessed dataset
df.to_csv('/content/drive/MyDrive/preprocessed_dataset.csv', index=False,encoding='utf-8-sig')
```
